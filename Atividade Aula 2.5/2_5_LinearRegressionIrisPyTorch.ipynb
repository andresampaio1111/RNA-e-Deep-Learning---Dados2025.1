# Regress√£o Linear com PyTorch

## Objetivo

Este exemplo utiliza regress√£o linear para estimar o comprimento das s√©palas da √≠ris a partir do seu comprimento das p√©talas.
Utiliza-se
- o c√°lculo do gradiente autom√°tico do PyTorch, por√©m
- a fun√ß√£o de perda e
- atualiza√ß√£o dos par√¢metros √© feita de forma expl√≠cita.
- A rede √© criada com uma camada nn.Linear()

## Importa√ß√£o dos pacotes

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn.datasets import load_iris

import torch
from torch import nn, optim
from torch.autograd import Variable

torch.manual_seed(1234)
```

## Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das s√©palas e p√©talas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normaliza√ß√£o dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias
```

```python
print('x_train_bias.shape:',x_train_bias.shape, x_train_bias.dtype)
```

```python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
print(y_train)
```

## Model

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinearNeuronio.png)

```python
#Inicializa√ß√£o usando toch.nn.Linear

model = torch.nn.Linear(in_features=2, out_features=1, bias=False)
```

### Verificando a inicializa√ß√£o dos par√¢metros

```python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight
```

```python
num_epochs = 100
learning_rate = 0.5
for epoch in range(num_epochs):
    #inputs = Variable(x_train_bias)
    #target = Variable(y_train)
    inputs = x_train_bias
    target = y_train

    # forward - predict
    y_pred = model(inputs)

    #loss c√°lculo da fun√ß√£o de perda
    loss = torch.mean((y_pred - target)**2)

    # backward propagation - c√°lculo gradient
    loss.backward()

    # gradiente descendente
    model.weight.data = model.weight.data - learning_rate * model.weight.grad.data

    # zeragem par√¢metros pr√≥ximo c√°lculo do gradiente
    model.weight.grad.data.zero_()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
```

# Exerc√≠cios

- 1) Por que no final da varredura do la√ßo de treinamento, a perda √© impressa utilizando
   `loss.data`? Por que precisa do `data`?

        Porque o loss √© um tensor conectado ao (mecanismo fundamental que permite ao PyTorch calcular automaticamente os gradientes), e o .data acessa apenas o valor num√©rico da perda, sem o hist√≥rico de gradientes.
        Isso √© necess√°rio para imprimir ou registrar o valor da perda sem interferir no c√°lculo dos gradientes.

        Contudo, nas vers√µes atuais do PyTorch, deve-se usar loss.item() ou loss.detach(), pois o uso de .data √© considerado obsoleto e inseguro.

- 2) Calcule e imprima o valor da perda (MSE) fazendo o predict da rede com os pesos j√°
   treinados.

```python
# üìò Avalia√ß√£o do modelo treinado - C√°lculo da perda (MSE)
# -------------------------------------------------------
# Agora usamos a rede j√° treinada para prever (sem autograd),
# e calculamos o erro m√©dio quadr√°tico (Mean Squared Error - MSE).

with torch.no_grad():  # desativa o autograd (n√£o h√° backprop)
    y_pred_final = model(x_train_bias)  # predi√ß√µes da rede
    mse_final = torch.mean((y_pred_final - y_train) ** 2)

print(f"Perda final (MSE) com pesos treinados: {mse_final.item():.6f}")
```

```python
# registrar ap√≥s atualizar os pesos

w0_cur, w1_cur = model.weight.detach().cpu().numpy().ravel().tolist()
w0_list.append(w0_cur)
w1_list.append(w1_cur)
```

```python
optimizer.step()

# registrar depois do step
with torch.no_grad():
    w0_cur, w1_cur = model.weight.detach().cpu().numpy().ravel().tolist()
w0_list.append(w0_cur)
w1_list.append(w1_cur)
```

```python
# ponto √≥timo anal√≠tico (est√°vel)

with torch.no_grad():
    # usa lstsq (est√°vel) para obter [w0; w1]
    w_opt, *_ = torch.linalg.lstsq(x_train_bias, y_train)  # shape (2,1)
    w0_opt, w1_opt = w_opt.squeeze().tolist()
```

```python
# Gr√°fico 2D da trajet√≥ria dos par√¢metros + ponto √≥timo anal√≠tico

import numpy as np
import matplotlib.pyplot as plt

# checagem r√°pida: precisamos de pelo menos 2 pontos (in√≠cio + 1 atualiza√ß√£o)
if len(w0_list) < 2:
    print("Trajet√≥ria vazia (len<2). Verifique se voc√™ est√° registrando ap√≥s cada atualiza√ß√£o de peso.")
else:
    w0_arr = np.array(w0_list, dtype=float)
    w1_arr = np.array(w1_list, dtype=float)

    plt.figure(figsize=(6,5))
    # pontos e linha da trajet√≥ria
    plt.scatter(w0_arr, w1_arr, s=28, label='Trajet√≥ria (√©pocas)')
    plt.plot(w0_arr, w1_arr, alpha=0.7)

    # marcar in√≠cio e fim
    plt.scatter([w0_arr[0]],  [w1_arr[0]],  s=90, marker='o', facecolors='none', edgecolors='tab:blue', label='In√≠cio')
    plt.scatter([w0_arr[-1]], [w1_arr[-1]], s=90, marker='s', facecolors='none', edgecolors='tab:green', label='Final (GD)')

    # marcar √≥timo anal√≠tico
    plt.scatter([w0_opt], [w1_opt], s=160, marker='*', color='gold', edgecolors='k', label='√ìtimo anal√≠tico')

    # garantir que n√£o ‚Äúcolapse‚Äù a escala quando a trajet√≥ria √© curta
    pad_x = max(1e-4, 0.05 * max(1.0, abs(w0_arr.max()-w0_arr.min())))
    pad_y = max(1e-4, 0.05 * max(1.0, abs(w1_arr.max()-w1_arr.min())))
    plt.xlim(w0_arr.min()-pad_x, w0_arr.max()+pad_x)
    plt.ylim(w1_arr.min()-pad_y, w1_arr.max()+pad_y)

    plt.title('Evolu√ß√£o dos par√¢metros (w0 √ó w1)')
    plt.xlabel('w0 (bias)')
    plt.ylabel('w1 (coeficiente)')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

    print(f"In√≠cio: w0={w0_arr[0]:.6f}, w1={w1_arr[0]:.6f}")
    print(f"Final (GD): w0={w0_arr[-1]:.6f}, w1={w1_arr[-1]:.6f}")
    print(f"√ìtimo anal√≠tico: w0={w0_opt:.6f}, w1={w1_opt:.6f}")
```
